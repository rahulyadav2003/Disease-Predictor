```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("./dataset/training_data.csv")
```

```python
df
```

```python
df.isna().sum().value_counts()
```

```python
df.columns
```

```python
df.drop('Unnamed: 133',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=24, random_state=72)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision = DecisionTreeClassifier().fit(X_train,y_train)
```

```python
decision.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random = RandomForestClassifier().fit(X_train,y_train)
```

```python
random.score(X_test,y_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(random.predict(X_test),y_test,average='micro')
```

```python
f1_score(decision.predict(X_test),y_test,average='micro')
```
